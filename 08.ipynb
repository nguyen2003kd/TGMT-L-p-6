```python
import matplotlib.pyplot as plt
```

```python
x=[50,30,80,10]
x1=[30,76,45,20]
y=[60,80,20,40]
y1=[70,30,59,60]
```

```python
plt.plot(x,y,'rs-',label='đường đỏ')
plt.plot(x1,y1,marker='o',color='b',linestyle='-',label='đường xanh')
plt.legend()
plt.xlabel('trục x')
plt.ylabel('trục y')
plt.show()
```

```python
categorise=['A','B','C','D','E','F']
values=[10,20,30,20,40,80]
```

```python
plt.bar(categorise, values, color='b', alpha=1) 
plt.xlabel('Danh mục') 
plt.ylabel('Giá trị') 
plt.show() 
```

```python
my_labels = ['A', 'B', 'C', 'D','E']
sizes = [15, 30, 25, 10,20]
my_colors = ['red', 'green', 'blue', 'yellow','white']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') # Vẽ biểu đồ hình tròn
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
x = [10, 20, 30, 40, 50]
y = [100, 200, 250, 300, 205]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')  
plt.title('Biểu đồ phân tán ví dụ') 
plt.xlabel('Trục x')  
plt.ylabel('Trục y') 
plt.legend()  
plt.show()  
```

```python
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)  
plt.title('Biểu đồ hộp ví dụ') 
plt.ylabel('Giá trị') 
plt.show() 
```

```python
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data)  
plt.title('Biểu đồ violin ví dụ')  
plt.ylabel('Giá trị')  
plt.show()  
```
